```python
# The charts are made using Plotly and Cufflinks. You need to set it up first for the code to run properly.
# https://plot.ly/ipython-notebooks/cufflinks/

# The code below has not been polished/optimized yet.

#Import libraries used
%matplotlib inline
#import pyfolio as pf
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
import pandas as pd

import pandas_datareader as pdr
import datetime
import numpy as np
import cufflinks as cf
#cf.go_offline()
cf.go_online()
import matplotlib.pyplot as plt
```

```python
# Analyze rolling returns

def rollingReturns(series, days):
    collector = []
    for i in range(len(series)):
        if i+days >= len(series):
            collector.append(np.NaN)
            continue
        collector.append((series.iloc[i+days,5]/series.iloc[i,5])-1)
    return collector

start = datetime.datetime(1993, 1, 1)
end = datetime.datetime(2017, 3, 3)

spy = pdr.get_data_yahoo('SPY',start=start, end=end)
spy["1yr rolling return"] = rollingReturns(spy,365)
spy["2yr rolling return"] = rollingReturns(spy,730)
spy["3yr rolling return"] = rollingReturns(spy,1095)
spy["4yr rolling return"] = rollingReturns(spy,1460)
spy["5yr rolling return"] = rollingReturns(spy,1825)
spy["6yr rolling return"] = rollingReturns(spy,2190)
spy
```

```python
## Make a rolling period returns chart using plotly

figure = spy[[str(x) + "yr rolling return" for  x in range(1,7)]].iplot(yTitle='Rolling Period Return', title='Analysis of Rolling Period Returns',
        subplots=True, subplot_titles=[str(x) + " Yr Rolling Return" for  x in range(1,7)], shape=(3,2),  asFigure=True)

#shared_yaxes=True,
for x in range(1,7):
    figure["layout"]["yaxis"+str(x)]["range"] = [-0.5,3]    
    
figure["layout"]["legend"]["orientation"] = "h"
#figure["layout"]["height"] =900 
#figure["layout"]["width"] =900


py.iplot(figure)
```

```python
# Calculating percentage of investing days are profitable after a number of years
def calculatePercentage(df):
    rolling_periods = [str(x) + "yr rolling return" for  x in range(1,7)]
    resultDict ={}
    for period in rolling_periods:
        
        series = df[period]
        maxgain = max(series)
        maxloss = min(series)
        ProportionNegativeReturns = len(series[series<=0])/series.count()
        fifthpercentile = series.quantile(q=0.05)
        twentyfifthpercentile = series.quantile(q=0.25)
        median = series.quantile(q=0.5)
        seventyfithpercentile = series.quantile(q=0.75)
        LessThanNegative10 = len(series[series<=-0.10])/series.count()
        MoreThanPositive10 = len(series[series>=0.10])/series.count()
        
     
        result = [maxgain,maxloss,ProportionNegativeReturns,fifthpercentile,twentyfifthpercentile,median,
                  seventyfithpercentile,LessThanNegative10,MoreThanPositive10]
        
        resultDict[period] = result
        
    resultDF = pd.DataFrame.from_dict(resultDict,orient="index")
    resultDF.columns = [
        "Max Gain", "Max Loss", "Proportion of Negative Returns", "5th Percentile","25th Percentile","Median",
        "75th Percentile", "Proportion of Less than -10% Returns","Proportion of More Than +10% Returns"
    ]
    return (resultDF)


        
ss = calculatePercentage(spy)
ss
```
